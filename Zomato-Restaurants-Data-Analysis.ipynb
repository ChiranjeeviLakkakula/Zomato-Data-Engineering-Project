Data Analysis

What is the country with the highest restaurants?

```python
data_zmt_ctry_count = data_zmt.groupby(data_zmt['country']).size().reset_index(name='counts').sort_values(by='counts',ascending=False)
```

```python
data_zmt_ctry_count['restaurant%'] = round((data_zmt_ctry_count['counts']/data_zmt_ctry_count['counts'].sum())*100,2)
```

```python
data_zmt_ctry_count
```

```python
data_zmt_ctry_count.loc[data_zmt_ctry_count['restaurant%']<3,'country'] = 'Other'
```

```python
data_zmt_ctry_count.groupby(['country']).sum().reset_index()
```

Countries with the Most Excellent-Rated Restaurants

```python
data_zmt['rating_text'].unique()
```

```python
data_zmt_excellent_ratings = data_zmt[data_zmt['rating_text']=='Excellent']
```

```python
data_zmt_excellent_ratings_cnt = data_zmt_excellent_ratings['country'].value_counts().reset_index(name='excellent_rating_counts')
```

```python
data_zmt_excellent_ratings_cnt['overall_rating%'] = round(data_zmt_excellent_ratings_cnt['excellent_rating_counts']/data_zmt_excellent_ratings_cnt['excellent_rating_counts'].sum(),2)
```

```python
data_zmt_excellent_ratings_cnt
```

Average Aggregate Rating of Restaurants by Country

```python
data_zmt.groupby(['country'])['aggregate_rating'].mean().reset_index(name='average_rating').sort_values(by='average_rating',ascending=False)
```

What are the most successful restaurants in India?

```python
data_zmt.columns
```

```python
#Indian Cities with the Most Restaurants

data_zmt_india = data_zmt[data_zmt['country']=='India']

data_zmt_indcity_rest = data_zmt_india.groupby(['city']).size().reset_index(name='no_of_restaurents').sort_values(by='no_of_restaurents',ascending=False)
```

```python
data_zmt_indcity_rest
```

```python
#Cities with the Most Successful Restaurants in India (High Rating & Votes)

data_zmt_succesful_rest = data_zmt[(data_zmt['aggregate_rating']>4) & (data_zmt['votes']>100)]

data_zmt_succesful_rest.groupby(['city']).size().reset_index(name='city_count').sort_values(by='city_count',ascending=False)
```

```python
data_zmt_rest_price = data_zmt.groupby(['price_range'])['votes'].sum().reset_index()

mapping = { 1: 'Economical',
            2: 'Mid-Range',
            3: 'Upscale',
            4: 'Fine Dining'}

data_zmt_rest_price['price_range'] = data_zmt_rest_price['price_range'].map(mapping)
```

```python
data_zmt_rest_price
```

Restaurants Price Analysis - Vote Distribution by Restaurant Price Range


```python
# Find the total number of votes for each price range

data_zmt_pricing_votes = data_zmt.groupby(['price_range'])['votes'].sum().reset_index(name='total_votes')

mapping = { 1: 'Economical',
            2: 'Mid-Range',
            3: 'Upscale',
            4: 'Fine Dining'}

data_zmt_pricing_votes['price_range'] = data_zmt_pricing_votes['price_range'].map(mapping)
```

```python
data_zmt_pricing_votes
```

Restaurants Price Analysis - Breakdown of Restaurants by Price Category

```python
data_zmt_price_range = data_zmt.groupby(['price_range']).size().reset_index(name='rest_price_range_cnt')

price_mapping = {
    1: 'Economical Eats',
    2: 'Mid-Range Dining',
    3: 'Upscale Cuisine',
    4: 'Fine Dining'
}

data_zmt_price_range['price_range'] = data_zmt_price_range['price_range'].map(price_mapping)
```

```python
data_zmt_price_range
```

Top 10 Restaurant Chains in India by Number of Outlets

```python
data_zmt_india = data_zmt[data_zmt['country'] == 'India']

data_zmt_rest_outlets  = data_zmt_india.groupby(['restaurant_name']).size().reset_index(name='rest_count').sort_values(by='rest_count',ascending=False)

data_zmt_top10_rest = data_zmt_rest_outlets.reset_index(drop=True).head(10)
```

Average Cost for Two at Top Indian Restaurant Chains (By Most Outlets)


```python
data_zmt_india_rest = data_zmt_india[['restaurant_name', 'average_cost_for_two']]

data_zmt_top10_restaurants = data_zmt_india_rest['restaurant_name'].value_counts().nlargest(10).index

data_zmt_india_rest[data_zmt_india_rest['restaurant_name'].isin(data_zmt_top10_restaurants)]
```
